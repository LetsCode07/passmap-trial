```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
import seaborn as sns
```

```python
df = pd.read_csv("pass.csv")
```

```python
df.head()
```

```python
#data is w.r.t to 100x100 field converting it to 120X80 to match with statsbomb mpl pitch
df['x']=df['x']*1.2
df['y']=df['y']*0.8
df['endX']=df['endX']*1.2
df['endY']=df['endY']*0.8
```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize=(13.5,8))
fig.set_facecolor('#22312b') 
ax.patch.set_facecolor('#22312b')
pitch = Pitch(pitch_color='grass', line_color='white', stripe=True)
pitch.draw(ax=ax)

plt.gca().invert_yaxis()


for i in range(len(df['x'])):
    if df['outcome'][i] == 'Successful':
        plt.plot((df['x'][i],df['endX'][i]), (df['y'][i],df['endY'][i]), color= 'blue')
        plt.scatter((df['x'][i],df['endX'][i]), (df['y'][i],df['endY'][i]), color= 'blue')
        
        
    if df['outcome'][i] == 'Unsuccessful':
        plt.plot((df['x'][i],df['endX'][i]), (df['y'][i],df['endY'][i]), color= 'red')
        plt.scatter((df['x'][i],df['endX'][i]), (df['y'][i],df['endY'][i]), color= 'red')
        
plt.title('Messi pass map vs Real Betis 2nd half', color = 'white')
```

```python
fig, ax = plt.subplots(figsize=(13.5,8))
fig.set_facecolor('#22312b') 
ax.patch.set_facecolor('#22312b')
pitch = Pitch(pitch_color='grass', line_color='white', stripe=True)
pitch.draw(ax=ax)

plt.gca().invert_yaxis()


for i in range(len(df['x'])):
    if df['outcome'][i] == 'Successful':
        dx = df['endX'][i] - df['x'][i]
        dy = df['endY'][i] - df['y'][i]
        x = df['x'][i]
        y= df['y'][i]
        plt.arrow(x,y,dx,dy,head_width=0.5,color='blue')
        
        
    if df['outcome'][i] == 'Unsuccessful':
        dx = df['endX'][i] - df['x'][i]
        dy = df['endY'][i] - df['y'][i]
        x = df['x'][i]
        y= df['y'][i]
        plt.arrow(x,y,dx,dy,head_width=0.5,color='red')
        
plt.title('Messi pass map vs Real Betis 2nd half', color = 'white')
```
